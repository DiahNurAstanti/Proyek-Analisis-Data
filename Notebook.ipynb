## Exploratory Data Analysis
### Explore Data Penyewaan Sepeda pada hari Libur dan Hari Kerja
Untuk menjawab pertanyaan #1, berikut tahapan yang dilakukan:
- Dataframe yang akan digunakan yaitu data perhari (day_df).
- Mula-mula persiapkan sebuah dataframe baru untuk menyimpan data penyewaan sepeda berdasarkan musimnya.
- Kemudian menambahkan kolom baru yang berisi nama dari tiap musim tersebut.

```python
season_df = day_df.groupby(by="season").agg({
    "cnt": "sum"
})
season_df.rename(columns={
    "cnt": "bike_rentals"
},inplace=True)
season_df["season_name"] = ["spring", "summer", "fall", "winter"]
season_df.head()
```

**Insight:**
- Jumlah penyewaan sepeda pada musim semi sebanyak 471348 Unit
- Jumlah penyewaan sepeda pada musim panas sebanyak 918589 Unit
- Jumlah penyewaan sepeda pada musim gugur sebanyak 1061129 Unit
- Jumlah penyewaan sepeda pada musim dingin sebanyak 841613 Unit


### Explore Data Penyewaan pada tahun 2012 berdasarkan bulan
Untuk menjawab pertanyaan #2, berikut tahapan yang dilakukan:
- Data yang digunakan yaitu data yang berasal dari dataframe day.
- Mula-mula siapkan sebuah dataframe baru untuk menyimpan data penjualan pada tahun 2012.
- Kemudian dataframe ini akan dikelompokkan berdasarkan bulannya serta menambahkan kolom baru yang berisi nama dari masing-masing bulan tersebut.

```python
on_2012_df = day_df[day_df.yr == 1]
on_2012_df.head ()
```

```python
month_2012_df = on_2012_df.groupby(by="mnth").agg({
    "cnt": "sum"
})
month_2012_df.rename(columns={
    "cnt": "bike_rentals"
}, inplace=True)
month_2012_df["month_name"] = ["January","February","March","April","May","June","July","August","September","October", "November","December"]
month_2012_df.head()
```

**Insight:**
- Puncak penyewaan sepeda terbanyak selama tahun 2012 terjadi pada bulan September.
- Penyewaan sepeda terendah selama tahun 2012 terjadi pada bulan Januari.

## Visualization & Explanatory Analysis
### Pertanyaan 1:
Data yang didapat untuk pertanyaan 1 akan disajikan dalam bentuk bar chart.

```python
plt.figure(figsize=(10,5))
colors = ["green", "yellow", "red", "skyblue"]
sns.barplot( 
    x="season_name", 
    y="bike_rentals",
    data= season_df, 
    palette= colors
)
plt.title("Bike Rentals By Season", fontsize = 22)
plt.xlabel(None)
plt.ylabel("Bike Rentals (Million)", fontsize = 10)
plt.xticks(fontsize = 15)
plt.show()
```

### Pertanyaan 2:
Data yang didapat untuk pertanyaan 2 akan disajikan dalam bentuk line chart.

```python
plt.figure (figsize=(15,5))
sns.lineplot(
    data=month_2012_df, 
    x="month_name",
    y="bike_rentals",
    linewidth = 2,
    marker = "o"
    
)
plt.xlabel(None)
plt.ylabel(None)
plt.title("Bike Rentals Performance in 2012", fontsize = 25)
plt.show()
```

## Conclusion
- Dari visualisasi data dengan bar chart, dapat diketahui dengan mudah bahwa penyewaan sepeda terbanyak terjadi pada musim gugur dan terendah pada musim semi.
- Berdasarkan line chart tersebut, didapatkan bahwa penyewaan sepeda pada tahun 2012 selalu mengalami penaikan dari awal tahun namun mengalami penurunan yang signifikan di akhir tahun. Diketahui juga, puncak tertinggi penyewaan sepeda terjadi pada bulan September dan penyewaan terendah terjadi pada bulan Januari.
